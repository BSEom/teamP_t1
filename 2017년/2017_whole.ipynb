```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
from matplotlib.font_manager import FontProperties

#    
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = FontProperties(fname=font_path)

# CSV  
q1 = pd.read_csv("2017_total_1q.csv")
q2 = pd.read_csv("2017_total_2q.csv")
q3 = pd.read_csv("2017_total_3q.csv")
q4 = pd.read_csv("2017_total_4q.csv")

#  
df_all = pd.concat([q1, q2, q3, q4], ignore_index=True)
df_all[""] = pd.to_datetime(df_all[""])

# 
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# 
sns.lineplot(data=df_all, x="", y="", hue="", marker="o")

#   ( FontProperties !)
plt.title("2017     ", fontsize=16, fontproperties=font_prop)
plt.xlabel("", fontproperties=font_prop)
plt.ylabel(" ()", fontproperties=font_prop)
plt.xticks(rotation=45, fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)
plt.legend(loc="best", fontsize=9, ncol=2, prop=font_prop)

#       
# plt.legend(
#     bbox_to_anchor=(1.05, 1),  #  
#     loc='upper left',
#     fontsize=9,
#     ncol=2,
#     prop=font_prop
# )
plt.tight_layout()
plt.show()


# df_all  
df_all.to_csv("2017_whole.csv", index=False, encoding="utf-8-sig")
```
