```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
from matplotlib.font_manager import FontProperties

# 📌 강제로 폰트 지정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = FontProperties(fname=font_path)

# CSV 파일 불러오기
q1 = pd.read_csv("2015_whole.csv")
q2 = pd.read_csv("2016_whole.csv")
q3 = pd.read_csv("2017_whole.csv")
q4 = pd.read_csv("2018_whole.csv")
q5 = pd.read_csv("2019_whole.csv")
q6 = pd.read_csv("2020_whole.csv")
q7 = pd.read_csv("2021_whole.csv")
q8 = pd.read_csv("2022_whole.csv")
q9 = pd.read_csv("2023_whole.csv")
q10 = pd.read_csv("2024_whole.csv")

# 하나로 합치기
df_all = pd.concat([q1, q2, q3, q4, q5, q6, q7, q8, q9, q10], ignore_index=True)
df_all["날짜"] = pd.to_datetime(df_all["날짜"])

# 시각화
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# 선그래프
sns.lineplot(data=df_all, x="날짜", y="최저가", hue="품목", marker="o")

# 그래프 꾸미기 (여기서 FontProperties 적용!)
plt.title("2015~2024년 부산 품목별 년도별 최저가 변화", fontsize=16, fontproperties=font_prop)
plt.xlabel("년도", fontproperties=font_prop)
plt.ylabel("최저가 (원)", fontproperties=font_prop)
plt.xticks(rotation=45, fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)
plt.legend(loc="best", fontsize=9, ncol=2, prop=font_prop)
plt.tight_layout()
plt.show()


# df_all을 저장하는 코드
df_all.to_csv("2015~2024_total.csv", index=False, encoding="utf-8-sig")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties

# 📌 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = FontProperties(fname=font_path)

# 연도별 파일 리스트
file_paths = [
    "2015_whole.csv", "2016_whole.csv", "2017_whole.csv",
    "2018_whole.csv", "2019_whole.csv", "2020_whole.csv",
    "2021_whole.csv", "2022_whole.csv", "2023_whole.csv",
    "2024_whole.csv"
]

# CSV 파일 읽기
df_all = pd.concat([pd.read_csv(f) for f in file_paths], ignore_index=True)
df_all["날짜"] = pd.to_datetime(df_all["날짜"])

# 📈 시각화
plt.figure(figsize=(16, 10))
sns.set(style="whitegrid")

plot = sns.lineplot(data=df_all, x="날짜", y="최저가", hue="품목", marker="o")

# 제목과 라벨
plt.title("2015~2024년 부산 품목별 연도별 최저가 변화", fontsize=16, fontproperties=font_prop)
plt.xlabel("년도", fontproperties=font_prop)
plt.ylabel("최저가 (원)", fontproperties=font_prop)
plt.xticks(rotation=45, fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)

# 🔍 범례 오른쪽으로 이동
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), fontsize=9, ncol=1, prop=font_prop)
plt.tight_layout()
plt.show()
```
