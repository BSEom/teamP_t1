```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
from matplotlib.font_manager import FontProperties

# 📌 강제로 폰트 지정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = FontProperties(fname=font_path)

# CSV 파일 불러오기
q1 = pd.read_csv("2015_total_1q.csv")
q2 = pd.read_csv("2015_total_2q.csv")
q3 = pd.read_csv("2015_total_3q.csv")
q4 = pd.read_csv("2015_total_4q.csv")

# 하나로 합치기
df_all = pd.concat([q1, q2, q3, q4], ignore_index=True)
df_all["날짜"] = pd.to_datetime(df_all["날짜"])

# 시각화
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# 선그래프
sns.lineplot(data=df_all, x="날짜", y="최저가", hue="품목", marker="o")

# 그래프 꾸미기 (여기서 FontProperties 적용!)
plt.title("2015년 부산 품목별 분기 최저가 변화", fontsize=16, fontproperties=font_prop)
plt.xlabel("날짜", fontproperties=font_prop)
plt.ylabel("최저가 (원)", fontproperties=font_prop)
plt.xticks(rotation=45, fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)
plt.legend(loc="best", fontsize=9, ncol=2, prop=font_prop)

# ⭐ 범례를 그래프 바깥 오른쪽 위로 배치
# plt.legend(
#     bbox_to_anchor=(1.05, 1),  # 오른쪽 바깥
#     loc='upper left',
#     fontsize=9,
#     ncol=2,
#     prop=font_prop
# )
plt.tight_layout()
plt.show()


# df_all을 저장하는 코드
df_all.to_csv("2015_whole.csv", index=False, encoding="utf-8-sig")
```
